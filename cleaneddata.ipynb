```python
# dependecies
# reference: fourth week's first day's ninth activity
import pandas as pd
from pathlib import Path
```

```python
# path
# reference: fourth week's first day's ninth activity
file = Path("Data/ufo_sighting_data.csv")
```

```python
# reading file
# reference: fourth week's first day's ninth activity
# xpert
data = pd.read_csv(file, encoding='utf-8')
```

```python
# verification
# reference: fourth week's first day's ninth activity
data.head()
```

```python
# cleaning for columns
# reference: fourth week's first day's ninth activity
columns = data[["Date_time",
                "city",
                "state/province",
                "country",
                "UFO_shape",
                "described_duration_of_encounter",
                "description",
                "date_documented",
                "latitude",
                "longitude"]]
columns.head()
```

```python
# renaming columns
# fourth week's first day's seventh activity
renamedcolumns = columns.rename(columns={"Date_time":"date_and_time",
                                          "city":"city2",
                                          "state/province":"state",
                                          "described_duration_of_encounter":"described_duration",
                                          "UFO_shape":"described_ufo",
                                          "description":"full_description"})
renamedcolumns.head()
```

```python
# cleaning for no nulls
# reference: fourth week's second day's third activity
nonulls = renamedcolumns.dropna(how='any')
nonulls.head()
```

```python
# cleaning for the u.s.
# xpert
us = nonulls[nonulls["country"] == "us"]
us.head()
```

```python
# deleting country
# fourth week's second day's third activity
del us["country"]
us.head()
```

```python
# splitting date_and_time
# xpert
us[["date", "time"]] = us["date_and_time"].str.split(" ", expand=True)
us.head()
```

```python
# deleting date_and_time
# fourth week's second day's third activity
del us["date_and_time"]
us.head()
```

```python
# organizing columns
# fourth week's first day's seventh activity
organizedus = us[["date_documented",
                      "date",
                      "time",
                      "described_duration",
                      "described_ufo",
                      "full_description",
                      "city2",
                      "state",
                      "latitude",
                      "longitude"]]
organizedus.head()
```

```python
# standard time
# xpert
organizedus["time"] = organizedus["time"].replace("24:00", "00:00")
organizedus["time"] = pd.to_datetime(organizedus["time"], format='%H:%M').dt.strftime('%I:%M %p')
organizedus.head()
```

```python
# replacing for punctuation
# xpert
punctuation = organizedus.replace({"&#44": ",",
                                   "&#33": "!",
                                   "&#39": "'",
                                   "&quot;": '"',
                                   "&#8216;":"'",
                                   "&#8217": "'",
                                   "&#8220": '"',
                                   "&#8221": '"',
                                   "&#8230": "...",
                                   "&#176": "°",
                                   "&#160": " ",
                                   "&#9": "    ",
                                   "&#8211": "-",
                                   "&#8212": "--",
                                   "&#188": "1/4",
                                   "&#186": "°",
                                   "&#8226;  ;": "•",
                                   ";": " ",
                                   "&#189;": "1/2"}, regex=True)
punctuation.head()
```

```python
# cleaning for cities_in_us
# xpert
splitcolumns = punctuation["city2"].str.split(r' \(', expand=True)
punctuation[["city", "parentheses"]] = splitcolumns[[0, 1]]
punctuation.head()
```

```python
# deleting city
# fourth week's second day's third activity
del punctuation["city2"]
punctuation.head()
```

```python
# deleting parentheses
# fourth week's second day's third activity
del punctuation["parentheses"]
punctuation.head()
```

```python
# organizing columns
# fourth week's first day's seventh activity
finishingtouches = punctuation[["date_documented",
                      "date",
                      "time",
                      "described_duration",
                      "described_ufo",
                      "full_description",
                      "city",
                      "state",
                      "latitude",
                      "longitude"]]
finishingtouches.head()
```

```python
# saving data
# fourth week's second day's third activity
finishingtouches.to_csv("Data/cleaneddata.csv", index=False, encoding='utf-8')
```
