```python
# dependecies
# reference: fourth week's first day's ninth activity
import pandas as pd
from pathlib import Path
```

```python
# path
# reference: fourth week's first day's ninth activity
file = Path("Data/us_cities_states_counties.csv")
```

```python
# reading file
# reference: fourth week's first day's ninth activity
# xpert
data = pd.read_csv(file, sep='|', encoding='utf-8')
```

```python
# verification
# reference: fourth week's first day's ninth activity
data.head()
```

```python
# cleaning for columns
# reference: fourth week's first day's ninth activity
columns = data[["City",
                "State short",
                "County"]]
columns.head()
```

```python
# renaming columns
# fourth week's first day's seventh activity
renamedcolumns = columns.rename(columns={"City":"city",
                                         "State short":"state",
                                         "County":"county"})
renamedcolumns.head()
```

```python
# cleaning for no nulls
# reference: fourth week's second day's third activity
nonulls = renamedcolumns.dropna(how='any')
nonulls.head()
```

```python
# cleaning for unique rows
# xpert
uniquerows = nonulls.drop_duplicates()
uniquerows.head()
```

```python
# lower case for merge
# xpert
uniquerows['city'] = uniquerows['city'].str.lower()
uniquerows['state'] = uniquerows['state'].str.lower()
uniquerows['county'] = uniquerows['county'].str.lower()
uniquerows.head()
```

```python
# saving data
# fourth week's second day's third activity
uniquerows.to_csv("Data/cleanedcounties.csv", index=False, encoding='utf-8')
```
