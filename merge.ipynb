```python
# dependency
# reference: fourth week's third day's first activity
import pandas as pd
```

```python
# dependency
# fourth week's first day's ninth activity
from pathlib import Path
```

```python
# paths
# fourth week's first day's ninth activity
firstfile = Path("Data/cleanedsightings.csv")
secondfile = Path("Data/cleanedcounties.csv")
```

```python
# reading files
# fourth week's first day's ninth activity
sightings = pd.read_csv(firstfile, encoding='utf-8')
counties = pd.read_csv(secondfile, encoding='utf-8')
```

```python
# verification
# fourth week's first day's ninth activity
sightings.head()
```

```python
# verification
# fourth week's first day's ninth activity
counties.head()
```

```python
# merging sightings and counties
# reference: fourth week's third day's first activity
# xpert
merge = pd.merge(sightings, counties, how='outer', on=['city', 'state'])
merge.head()
```

```python
# cleaning for no nulls
# reference: fourth week's second day's third activity
nonulls = merge.dropna(how='any')
nonulls.head()
```

```python
# organizing columns
# fourth week's first day's seventh activity
organizedmerge = nonulls[["date_documented",
                          "date",
                          "time",
                          "described_duration",
                          "described_ufo",
                          "full_description",
                          "city",
                          "county",
                          "state",
                          "latitude",
                          "longitude"]]
organizedmerge.head()
```

```python
# saving data
# fourth week's second day's third activity
organizedmerge.to_csv("Data/merge.csv", index=False, encoding='utf-8')
```
