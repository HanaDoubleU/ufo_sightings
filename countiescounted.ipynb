```python
# dependency
# reference: fourth week's third day's first activity
import pandas as pd
```

```python
# dependency
# fourth week's first day's ninth activity
from pathlib import Path
```

```python
# path
# fourth week's first day's ninth activity
file = Path("Data/merge.csv")
```

```python
# reading files
# fourth week's first day's ninth activity
merge = pd.read_csv(file, encoding='utf-8')
```

```python
# verification
# fourth week's first day's ninth activity
merge.head()
```

```python
# countiescounted
# xpert
countiescounted = merge[['county', 'state']].value_counts().reset_index()
countiescounted.columns = ['county', 'state', 'counts']
countiescounted.head()
```

```python
# saving data
# fourth week's second day's third activity
countiescounted.to_csv("Data/countiescounted.csv", index=False, encoding='utf-8')
```
