```python
# dependencies
# reference: fourth week's third day's first activity
# xpert
import pandas as pd
import geopandas as gpd
```

```python
# dependency
# fourth week's first day's ninth activity
from pathlib import Path
```

```python
# path
# fourth week's first day's ninth activity
file = Path("Data/US_County_Boundaries.geojson")
```

```python
# reading files
# fourth week's first day's ninth activity
geodata = gpd.read_file(file)
```

```python
# verification
# fourth week's first day's ninth activity
geodata.head()
```

```python
# cleaning for columns
# reference: fourth week's first day's ninth activity
columns = geodata[["NAME",
                   "STATE_NAME",
                   "geometry"]]
columns.head()
```

```python
# lower case for merge
# xpert
columns['NAME'] = columns['NAME'].str.lower()
columns['STATE_NAME'] = columns['STATE_NAME'].str.lower()
columns.head()
```

```python
# renaming columns
# fourth week's first day's seventh activity
renamedcolumns = columns.rename(columns={"NAME":"county",
                                         "STATE_NAME":"state"})
renamedcolumns.head()
```

```python
# saving data
# fourth week's second day's third activity
# xpert
renamedcolumns.to_file("Data/cleanedgeodata.geojson", driver='GeoJSON')
```
